3.Remove rows with more than 5 missing value

```python
df=data[data.isnull().sum(axis=1) < 5]
df
```

4.Fill missing values in age column with mean of age


```python
mean=data["Age"].mean()
data["Age"]=data["Age"].fillna(mean)
data["Age"]
```

5. Fill missing values in fare column with average of fare using simpleimputer

```python
Fare_mean=data["Fare"].mean()
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean', fill_value=Fare_mean)
imputer = imputer.fit(data[["Fare"]])
data["Fare"] = imputer.transform(data[["Fare"]])
data["Fare"]
```
