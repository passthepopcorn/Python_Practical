```python
import numpy as np
import pandas as pd
data=pd.read_csv('/content/IMDB-Movie-Data.csv')
data
```

```python
#1.Groupby operations: List the average rating for each director
mean=data[["Director","Rating"]].groupby(["Director","Rating"]).mean()
mean
```

2.Sorting operation: List the average rating for each ‘Director’ and sort them from highly
rated to lowest

```python
mean.sort_values(by=['Director','Rating']).mean
```

```python
#3.Dealing with missing values
data.isna().sum()
```

```python
#4.Dropping columns and null values:
#a. Drop the Metascore’ column completely from the data. The operation should make changes to original data
d=data.drop(axis=1,columns='Metascore')
d
```

```python
dro=data.isna().sum()>5
dro
```

```python
#c. Fill these null values of Revenue(Millions)with mean Revenue (Millions).
mean=data['Revenue (Millions)'].mean()
data['Revenue (Millions)']=data['Revenue (Millions)'].fillna(mean)
data.isna().sum()
```

```python
#5.apply( ) functions:
#d. Write a function to classify the movies based on their ratings as good(>=7.5), average(>=5) bad(<5),
def avgCal(rating):
  if(rating>=7.5):
    return "Movie is Good"
  elif(rating<7.5) & (rating>=5):
    return "Movie Average"
  elif(rating<=5):
    return "Movie is Bad"
print(avgCal(2.5))
```

```python
#e. Apply this function to the dataframe’s Rating column.
data['Rating'].apply(avgCal)
data
```
